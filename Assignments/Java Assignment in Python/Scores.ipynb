```python
import pandas as pd
```

```python
df = pd.read_csv('Scores.csv', sep=',|-|_', names=['Player', 'Against_1', 'Runs_1', 'Day_1', 'Month_1', 'Against_2', 'Runs_2', 'Day_2', 'Month_2', 'Against_3', 'Runs_3', 'Day_3', 'Month_3', 'Against_4', 'Runs_4', 'Day_4', 'Month_4', 'Against_5', 'Runs_5', 'Day_5', 'Month_5'], header=None)
df.head()
```

```python
df['max_run'] = df[['Runs_1', 'Runs_2', 'Runs_3', 'Runs_4', 'Runs_5']].max(axis=1)
df.groupby('Player').max()[['max_run']]
```
