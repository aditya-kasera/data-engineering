```python
god = pd.read_csv('greek_gods.csv')
god
```

1.Merge the data from greek_gods.csv and greek_goddesses.csv based on a common field and create a new table that includes information about both gods and goddesses.

```python
merged = pd.concat([god, goddess])
merged = merged.reset_index(drop=True)
merged
```

2.Filter the merged table to only include gods and goddesses who are older than 8000 years, then sort them based on their ages in descending order.

```python
merged[merged['Age']>8000].sort_values(by='Age',ascending=False)
```

3.Join the two tables based on the "Domain" field and calculate the average age of gods and goddesses in each domain.

```python
joined = pd.merge(god, goddess, on='Domain', how='outer')
joined
```

```python
average_age = joined.groupby('Domain')[['Age_x', 'Age_y']].mean()
average_age['Average Age'] = average_age.mean(axis=1)
average_age = average_age.rename(columns={'Age_x': 'Gods Age', 'Age_y': 'Goddess Age'})
average_age
```

4.Determine which god/goddess has the highest age, and then find out if they are a god or a goddess.

```python
max_age = average_age['Average Age'].max()

if max_age in average_age['Gods Age'].values:
    name = god.loc[god['Age'] == max_age, 'God']
    print(" God's maximum age = ", max_age, "\n", name)
    
elif max_age in average_age['Goddess Age'].values:
    name = goddess.loc[goddess['Age'] == max_age, 'Goddess']
    print(" Goddess has maximum age of ", max_age, "\n", name)
```


5.Create a new column in each table called "Age_Group" and categorize the gods/goddesses into groups such as "Young" (age < 5000), "Middle-aged" (age between 5000 and 8000), and "Old" (age > 8000).


```python
def categorize(age):
    if age < 5000:
        return "Young"
    elif 5000 <= age <= 8000:
        return "Middle-aged"
    elif age > 8000:
        return "Old" 

god['Age_Group'] = god['Age'].apply(categorize)
god
```

```python
goddess['Age_Group'] = god['Age'].apply(categorize)
goddess
```


6.Compare the average ages of gods and goddesses. Is there a significant age difference between them? If yes, which group tends to be older?


```python
avg_god = god['Age'].mean()
avg_goddess = goddess['Age'].mean()

if avg_god > avg_goddess:
    print("Gods are older by ", (avg_god - avg_goddess), " years.")
elif avg_god < avg_goddess:
    print("Goddesses are older by ", (avg_goddess - avg_god), " years.")
else:
    print("Both are of same age")
```
